```python
from data import get_dataset
import torch.utils.data as Data
import os
import pickle
import networkx as nx

datasets = ["actor","chameleon","citeseer","community","computer","cora","cornell","cycle","grid","photo","pubmed","shape","squirrel","texas","wiki","wisconsin"]
for data in datasets:
    graph, features, labels, idx_train, idx_val, idx_test = get_dataset(data, 50, 25, 0)
    graph = graph.to_networkx().to_undirected()
    k = sum(dict(graph.degree()).values()) / len(graph.nodes())
    count = 0
    edges1 = [(u,v) for u,v in graph.edges() if u<v]
    edges2 = [(v,u) for u,v in graph.edges() if u>v]
    edges = list(set(edges1+edges2))
    for u,v in edges:
        if labels[u] == labels[v]:
            count += 1
    homophily = count/len(edges)
    print(data, len(graph.nodes()), len(graph.edges()), k, features.shape[1],labels.max().item() + 1, homophily)
    print(features.shape)
    
```

```python
from data import get_dataset
import torch.utils.data as Data
import os
import pickle
import networkx as nx
import dgl

datasets = ["actor","chameleon","citeseer","community","computer","cora","cornell","cycle","grid","photo","pubmed","shape","squirrel","texas","wiki","wisconsin"]
for data in datasets:
    graph, features, labels, idx_train, idx_val, idx_test = get_dataset(data, 50, 25, 0)
    graph = dgl.to_bidirected(graph)
    homophily = dgl.node_homophily(graph, labels)
    graph = graph.to_networkx().to_undirected()
    k = sum(dict(graph.degree()).values()) / len(graph.nodes())
    print(data, len(graph.nodes()), len(graph.edges()), k, features.shape[1],labels.max().item() + 1, homophily)
    print(features.shape)
```
